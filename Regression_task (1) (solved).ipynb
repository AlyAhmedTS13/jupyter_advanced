# Implementation

# **This an implementation of the mean squared loss and mean absolute loss use either of them**

**There are also implementation of the l2_cost and l1_cost**

**Below is the mathmatical proof for mean squared error loss if you like you can try to prove the mean absolute error.**
\begin{align}
        \mathbf{loss}= 1/2M\sum_{i}^n (pred- label)^2
    \end{align}
        pred = w*x+b,
    M = number samples ,
    label is our Y
\begin{align}
        \frac{\partial loss}{\partial w}=1/M\sum_{i}^n ((wx+b)- label) \frac{\partial }{\partial w}(wx+b-label)
    \end{align}
\begin{align}
      \frac{\partial loss}{\partial w}= 1/M\sum_{i}^n (wx+b- label)x
      \end{align}
\begin{align}
        \frac{\partial loss}{\partial b}=1/M\sum_{i}^n ((wx+b)- label) \frac{\partial }{\partial b}(wx+b-label)
    \end{align}
\begin{align}
      \frac{\partial loss}{\partial b}= 1/M\sum_{i}^n (wx+b- label)
      \end{align}   

```python
def compute_gradient(x, y, w, b):
  # x-->featured
  # y-->labels
  # w--> initial w
  # b-> initial b
###### Write your code here #############
    M = len(y)
    pred = w * x + b 
    dj_dw = (1 / M) * sum((pred - y) * x)
    dj_db = (1 / M) * sum(pred - y)

    return dj_dw, dj_db
```



> the target of the function is to compute and return \begin{align} \frac{\partial loss}{\partial w}
,\frac{\partial loss}{\partial b}
  \end{align}

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iters):
  # x-->features
  # y-->labels
  # w_in--> initial w
  # b_in> initial b
  # alpha-->alpha
  # num_iters--> number of iterations
  #compute_gradient--> the previous function that returns dj_dw, dj_db

###### Write your code here #############
    
    for i in range(num_iters):
        
    
        dj_dw, dj_db=compute_gradient(x, y, w, b)
    
        w_in = w_in - alpha * dj_dw
    
        b_in =b_in - alpha * dj_db







    return w_in, b_in
```



> this function call compute_gradient at every iteration to get the
\begin{align} \frac{\partial loss}{\partial w}
,\frac{\partial loss}{\partial b}
  \end{align}
  and then compute new (w and b)
simultaneously



```python
def compute_model_output(x, w, b):
#x--> features
#w-->final w
#b-->final b
###### Write your code here #############


    y_predicted = w * x + b


    return y_predicted
```



> the target of the function is to compute the prediction for every instance in the data and return them



```python
def compute_cost(x, y, w, b):
  # x-->features
  # y-->labels
  # w--> final w
  # b--> final b
###### Write your code here #############
    M = len(y)


    y_predicted = compute_model_output(x, w, b)

    total_cost = (1 / (2 * M)) * sum((y_predicted - y) ** 2)

    return total_cost
```



> the target of the function is to compute the loss(cost function) and return it



# Modeling

```python
w_init = 5
b_init = -1
iterations =5000 # choose the suitable number of iterations
tmp_alpha = 1e-3

w_final, b_final = gradient_descent(x_train ,y_train, w_init, b_init, tmp_alpha,
                                                    iterations)

print(f'final(w):{w_final} , final(b):{b_final}')

print(f'minimum cost:{compute_cost(x_train, y_train, w_final, b_final)}')

Y_predicted = compute_model_output(x_train, w_final, b_final)

plt.plot(x_train, Y_predicted, c='b',label='Our Prediction')
plt.scatter(x_train,y_train,c='r',label='Actual values')
plt.title('Regression')
plt.ylabel('Labels')
plt.xlabel('Features')
plt.legend()
plt.show()
```
